```python
import pyspark

from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.getOrCreate()
```

```python
data = [["1", "sravan", "IT", 45000],
        ["2", "ojaswi", "CS", 85000],
        ["3", "rohith", "CS", 41000],
        ["4", "sridevi", "IT", 56000],
        ["5", "bobby", "ECE", 45000],
        ["6", "gayatri", "ECE", 49000],
        ["7", "gnanesh", "CS", 45000],
        ["8", "bhanu", "Mech", 21000]
        ]

# specify column names
columns = ['ID', 'NAME', 'DEPT', 'FEE']

# creating a dataframe from the lists of data
dataframe = spark.createDataFrame(data, columns)
```

```python
dataframe.show()
dataframe.printSchema()
```

```python
from pyspark.sql.functions import col,lit
```

```python
dataframe.withColumn('FEE',col('FEE')*2).show()
```

```python
dataframe = dataframe.withColumn("ID", col("ID").cast("integer"))
```

```python
dataframe.printSchema()
```

```python
df=dataframe.withColumn("collage",lit("UMKC"))
df.show()
```

```python
df=df.withColumnRenamed('collage','university')
```

```python
df.show()
```
